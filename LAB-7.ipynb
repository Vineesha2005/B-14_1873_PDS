```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset (You can replace with actual CSV file path)
data = {
    "Year": [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020],
    "Population": [2.5, 3.0, 3.7, 4.4, 5.3, 6.1, 6.9, 7.8]  # in billions
}
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(8,5))
plt.plot(df["Year"], df["Population"], marker='o', linestyle='-', color='b')
plt.xlabel("Year")
plt.ylabel("World Population (Billions)")
plt.title("World Population Growth Over Years")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (replace with "AAPL.csv")
# Example structure: Date, Close
data = {
    "Date": pd.date_range(start="2020-01-01", periods=10, freq="M"),
    "Close": [75, 80, 85, 82, 90, 95, 100, 110, 105, 120]
}
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(10,5))
plt.plot(df["Date"], df["Close"], marker='o', linestyle='-', color='g')
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.title("Apple Stock Price Trend")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
