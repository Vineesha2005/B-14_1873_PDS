```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("student_marks.csv")

# Plot bar chart
plt.figure(figsize=(8,5))
plt.bar(df["Subject"], df["Average_Marks"], color=["blue", "green", "red", "purple", "orange"])
plt.title("Average Marks of Students in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()

# Insights
max_sub = df.loc[df["Average_Marks"].idxmax(), "Subject"]
min_sub = df.loc[df["Average_Marks"].idxmin(), "Subject"]
print(f"Easiest Subject: {max_sub} ({df['Average_Marks'].max()})")
print(f"Toughest Subject: {min_sub} ({df['Average_Marks'].min()})")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("sales_category.csv")

# Plot bar chart
plt.figure(figsize=(8,5))
plt.bar(df["Category"], df["Sales"], color=["cyan", "orange", "green", "red", "purple"])
plt.title("Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Sales Amount")
plt.show()

# Insights
best_cat = df.loc[df["Sales"].idxmax(), "Category"]
print(f"Best-Selling Category: {best_cat}")
```
